Gotcha! The lemmatizer is actually pretty complicated, it needs Parts of Speech (POS) tags.

```python
import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer

nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

hamlet_text = gutenberg.raw('shakespeare-hamlet.txt')

hamlet_words = word_tokenize(hamlet_text)

nltk.download('averaged_perceptron_tagger_eng')

tagged_words = pos_tag(hamlet_words)

wnl = WordNetLemmatizer()

def penn2morphy(penntag):
    """Converts Penn Treebank tags to WordNet POS tags."""
    morphy_tag = {'NN': 'n', 'JJ': 'a', 'VB': 'v', 'RB': 'r'}
    try:
        return morphy_tag[penntag[:2]]
    except KeyError:
        return 'n'

lemmatized_words = [
    wnl.lemmatize(word, pos=penn2morphy(tag)) for word, tag in tagged_words
]

print(lemmatized_words[:50])
```

```python
# Lemmatize tagged words using the correct WordNet POS tag and lowercase the words
lemmatized_words = [
    wnl.lemmatize(word.lower(), pos=penn2morphy(tag)) for word, tag in tagged_words
]

print(lemmatized_words[:50])
```

Now, lets create a new lemmatization function for sentences given what we learnt above.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer

# Initialize WordNet Lemmatizer
wnl = WordNetLemmatizer()

def penn2morphy(penntag):
    """ Converts Penn Treebank tags to WordNet POS tags. """
    morphy_tag = {'NN': 'n', 'JJ': 'a', 'VB': 'v', 'RB': 'r'}
    try:
        return morphy_tag[penntag[:2]]
    except KeyError:
        return 'n'

# Lemmatization function
def lemmatize_sent(text):
    """
    Lemmatizes the sentence text.
    Returns a list of lowercased lemmatized words.
    """
    return [wnl.lemmatize(word.lower(), pos=penn2morphy(tag))
            for word, tag in pos_tag(word_tokenize(text))]

# Example sentence
lemmatized_words = lemmatize_sent('He is walking to school')
print(lemmatized_words)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()

def penn2morphy(penntag):
    """ Converts Penn Treebank tags to WordNet POS tags. """
    morphy_tag = {'NN': 'n', 'JJ': 'a', 'VB': 'v', 'RB': 'r'}
    try:
        return morphy_tag[penntag[:2]]
    except KeyError:
        return 'n'

def lemmatize_sent(text):
    """
    Lemmatizes the sentence text.
    Returns a list of lowercased lemmatized words.
    """
    return [wnl.lemmatize(word.lower(), pos=penn2morphy(tag))
            for word, tag in pos_tag(word_tokenize(text))]

# Example sentence
lemmatized_words = lemmatize_sent('long liue the king')
print(lemmatized_words)
```

Before we test our classifier on the test set, we get a sense of how good it is on the validation set.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder
import pandas as pd

def preprocess_text(text):
    return text.split()

hamlet_text = """ACT I. SCENE I. Elsinore. A platform before the castle.
FRANCISCO at his post. Enter to him BERNARDO.
BER. Who's there?
FRAN. Nay, answer me: stand, and unfold yourself.
BER. Long live the king!
FRAN. Bernardo?
BER. He.
FRAN. You come most carefully upon your hour.
BER. 'Tis now struck twelve; get thee to bed, Francisco.
FRAN. For this relief much thanks: 'tis bitter cold,
And I am sick at heart.
BER. Have you had quiet guard?
FRAN. Not a mouse stirring.
BER. Well, good night.
If you do meet Horatio and Marcellus,
The rivals of my watch, bid them make haste.
"""

df = pd.DataFrame({
    'Text': [hamlet_text[:500], hamlet_text[500:1000], hamlet_text[1000:1500]],
    'Title': ["Hamlet, Act 1, Scene 1", "Hamlet, Act 1, Scene 2", "Hamlet, Act 1, Scene 3"]
})

count_vect = CountVectorizer(analyzer=preprocess_text)

count_vect.fit(df['Text'])

train_set, valid_set, train_tags, valid_tags = train_test_split(
    df['Text'], df['Title'], test_size=0.2, random_state=42
)

all_tags = pd.concat([train_tags, valid_tags])

label_encoder = LabelEncoder()
label_encoder.fit(all_tags)

train_tags_encoded = label_encoder.transform(train_tags)
valid_tags_encoded = label_encoder.transform(valid_tags)

train_set = count_vect.transform(train_set)
valid_set = count_vect.transform(valid_set)

# Debug: Check shape of transformed data and labels
print("Train set shape:", train_set.shape)
print("Valid set shape:", valid_set.shape)
print("Train tags encoded:", train_tags_encoded)
print("Valid tags encoded:", valid_tags_encoded)

clf = MultinomialNB()
clf.fit(train_set, train_tags_encoded)

predictions_valid = clf.predict(valid_set)

print("Predictions:", predictions_valid)

accuracy = accuracy_score(predictions_valid, valid_tags_encoded)
print('Dataset accuracy = {}'.format(accuracy * 100))
```
